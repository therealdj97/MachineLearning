LR performs the task to predict a dependent variable value(y) based on given independent variable(x)

x=input training data(univariate -one input variable)
y=labels the data (supervised learning)

```python
import pandas as pd
import numpy as nm
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv("homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel("area")
plt.ylabel("prices")
plt.scatter(df.area,df.price,color='red',marker="*")
```

```python
new_df=df.drop('price',axis='columns')
new_df
```

```python
price=df.price
price
```

```python
#create linear regression object
reg=linear_model.LinearRegression()
reg.fit(new_df,price)
```

```python
#**predict price of home with area = 3300sqr ft**
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
