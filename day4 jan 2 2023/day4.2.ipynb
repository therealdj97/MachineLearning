```python
import pandas as pd
import numpy as nm
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices (1).csv")
df
```

data processing fill na values with median value of column

```python
df.bedrooms.median()
```

```python
df.bedrooms= df.bedrooms.fillna(df.bedrooms.median())
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df.drop('price',axis='columns'),df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]])
```

```python
#predicting price of house with 2500sq ft 4 bed rooms 5 yea old
reg.predict([[2500,4,5]])
```
