```python
import pandas as pd
```

```python
df=pd.read_csv("titanic.csv")
```

```python
df
```

```python
inputs = df.drop('Survived',axis='columns')
```

```python
target =df['Survived']
```

```python
from sklearn.preprocessing import LabelEncoder
le_Pclass = LabelEncoder()
le_Sex = LabelEncoder()
le_Age = LabelEncoder()
le_Fare = LabelEncoder()
```

```python
inputs['Pclass_n'] = le_Pclass.fit_transform(inputs['Pclass'])
inputs['Sex_n'] = le_Sex.fit_transform(inputs['Sex'])
inputs['Age_n'] = le_Age.fit_transform(inputs['Age'])
inputs['Fare_n'] = le_Fare.fit_transform(inputs['Fare'])
```

```python
inputs
```

```python
inputs = inputs.drop(['Pclass','Age','Sex','Fare','PassengerId','Name','Ticket','Cabin','Embarked','SibSp','Parch'],axis='columns')
```

```python
inputs
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs,target)
```

```python
model.score(inputs,target)
```

```python
model.predict([[2,0,88,131]])
```
