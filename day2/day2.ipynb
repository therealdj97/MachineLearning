# Real-world datasets often contain features that are varying in degrees of magnitude, range and units. Therefore, in order for machine learning models to interpret these features on the same scale, we need to perform feature scaling.

# min max scaller is used for skewed data also when your dataset has high veriance 

The machine learning algorithm works on numbers and does not know what that number represents. A weight of 10 grams and a price of 10 dollars represents completely two different things â€” which is a no brainer for humans, but for a model as a feature, it treats both as same.

there are 2 methods two achieve featue scaling 
1.   Min max scaller (0 to 1)
2.   Standard Scaler (-1 to 1)

if we have -ve values in dataset and we apply min max scaller we will get same result as of std scaller

```python
# Min max scaller
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df = pd.DataFrame({'WEIGHT': [15, 18, 12,10],
                   'PRICE': [1,3,2,5]},
                   index = ['Orange','Apple','Banana','Grape'])
df1 = pd.DataFrame(scaler.fit_transform(df),
                   columns=['WEIGHT','PRICE'],
                   index = ['Orange','Apple','Banana','Grape'])

ax = df.plot.scatter(x='WEIGHT', y='PRICE',color=['red','green','blue','yellow'], 
                     marker = '*',s=80, label='BREFORE SCALING');
df1.plot.scatter(x='WEIGHT', y='PRICE', color=['red','green','blue','yellow'],
                 marker = 'o',s=60,label='AFTER SCALING', ax = ax);
plt.axhline(0, color='red',alpha=0.2)
plt.axvline(0, color='red',alpha=0.2);
```

```python
# Min max scaller
df=pd.read_csv("Iris.csv")
df.drop("Species",axis=1,inplace=True)
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df1=pd.DataFrame(scaler.fit_transform(df),columns=['Id', 'SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'])
ax=df.plot.scatter(x="SepalLengthCm",y="SepalWidthCm",marker="*",label="Before-Scaling",color="red")
df1.plot.scatter(x="SepalLengthCm",y="SepalWidthCm",marker="+",label="After-Scaling",ax=ax)
plt.axhline(0, color='red',alpha=0.2)
plt.axvline(0, color='red',alpha=0.2);
plt.show()
```

2.   Manhatten dist

Euclidian dist. represents shortest dist between 2 points |
can be used for k-means

```python
from scipy.spatial import distance
import math

p1 =(1,2,3)
p2 = (4,5,6)

math.dist(p1,p2)
```

Manhatten dist is sum of absolute dist between 2 points across all dimentions
dist = |x1-x2| + |y1-y2|

```python
def get_manhatten_dist(p,q):
  """
    will give manhatten dist between 2 points assuming they are of same dim
  """
  dist=0
  for p_i,q_i in zip(p,q):
    dist += abs(p_i - q_i)
  return dist
```

```python
a=(1,1)
b=(4,3)
# dist between a and b
d= get_manhatten_dist(a,b)
print(d)
```

Principle componenet analysis
# it is widely used tool fopr EDA in ML for predictive models
PCA is an unsupervised statastical technique used to examine interrelations among set of variables
also known as general factor analysis where regression determines a line of best fit 

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
df = pd.DataFrame(cancer['data'], columns = cancer['feature_names'])
df
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
scalar.fit(df)
scaled_data = scalar.transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
pca.fit(scaled_data)
x_pca = pca.transform(scaled_data)
x_pca.shape
#transformed data into std scaler furthermore  transformed it into pca scaller 
```

```python
plt.figure(figsize = (8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c= cancer['target'],cmap="plasma")
plt.xlabel("First PC")
plt.ylabel("Second PC")
```

```python
pca.components_
```

```python
df_comp =pd.DataFrame(pca.components_, columns=cancer['feature_names'])
plt.figure(figsize =(14,6))
sns.heatmap(df_comp)
```
