# Hierarchical Clustering

```python
unsupervised learning method
this algorithm builds clusters by measuring dissimilarities between data
this method can be used on any data to interpret/visualize relationship between individual data points
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [4,5,10,4,3,11,14,6,10,12]
y = [21,19,24,17,16,25,24,22,21,21]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
#Now we can compute the ward linkage using euclidian distance and visualize it using a dendogram   
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
x = [4,5,10,4,3,11,14,6,10,12]
y = [21,19,24,17,16,25,24,22,21,21]

data = list(zip(x,y))

linkage_data = linkage(data, method = 'ward', metric = 'euclidean')
dendrogram(linkage_data)

plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
x = [4,5,10,4,3,11,14,6,10,12]
y = [21,19,24,17,16,25,24,22,21,21]

data = list(zip(x,y))

hierarchical_cluster = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean', linkage = 'ward')
labels = hierarchical_cluster.fit_predict(data)

plt.scatter(x,y,c = labels)
plt.show()
```

# K Means Clustering Algorithm

```python
Unsupervised
Divides data into K clusters by minimizing the variance in each cluster
Estimate best value of K using elbow method - then use k-means clustering to group the data points into clusters
```

```python
import matplotlib.pyplot as plt

x = [4,5,10,4,3,11,14,6,10,12]
y = [21,19,24,17,16,25,24,22,21,21]

plt.scatter(x,y)
plt.show()
```

```python
from sklearn.cluster import KMeans

data = list(zip(x,y))
inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)
    
plt.plot(range(1,11), inertias, marker = 'o')
plt.title("Elbow method")
plt.xlabel("Number of clusters")
plt.ylabel("Intertia")
plt.show()
```

```python
kmeans = KMeans(n_clusters = 2)
kmeans.fit(data)

plt.scatter(x,y, c=kmeans.labels_)
plt.show()
```
